Function to get the coodinates

```python
import geocoder # import geocoder

def getCoordinates(code):
    # initialize your variable to None
    lat_lng_coords = None
    count = 0;
    # loop until you get the coordinates
    while(lat_lng_coords is None):
        count = count+1
        if count > 10:
            print('More than 10 attempts')
            return 0,0
        g = geocoder.google('Downtown Toronto, Toronto, Ontario')
        lat_lng_coords = g.latlng
    latitude = lat_lng_coords[0]
    longitude = lat_lng_coords[1]
    return latitude,longitude
```

Iterate through all the postcodes

```python
lats=[]
longs=[]
for i, row in df_post.iterrows():
    print(row['Postcode'])
    code = row['Postcode']
    latitude,longitude = getCoordinates(code)
    if latitude==0:
        lats.append(0)
        longs.append(0)
    else:
        lats.append(latitude)
        row.append(longitude)
```

Geocode is working, but it acept only the postcode in the query string.

```python
import pgeocode
nomi = pgeocode.Nominatim('ca')
lats=[]
longs=[]
for i, row in df_post.iterrows():
    print(row['Postcode'])
    code = row['Postcode']
    location = nomi.query_postal_code(code)
    print(location.latitude, location.longitude)
    lats.append(location.latitude)
    longs.append(location.longitude)
```

Insert the latitudes and longitudes in the df_post

```python
df_post['Latitude']=lats
df_post['Longitude']=longs
```
